```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
from sklearn.preprocessing import StandardScaler
driverdata = pd.read_csv("D:\\1_Lecture Material\\Term_2\\Pedram Habibi_Business_Processing\\Driver-data.csv")
dd=StandardScaler().fit_transform(driverdata)
#dd=driverdata
```

```python
data=pd.DataFrame(dd)
```

```python
data.columns = [ 'id','mean_dist_day', 'mean_over_speed_perc']
dd=data.drop(columns=['id'])
model = KMeans (n_clusters=4)
model.fit(dd)
print(model.labels_)
```

```python
print(driverdata.head())
```

```python
colormap = np.array(['red', 'blue','green','black'])
plt.scatter(driverdata.mean_dist_day, driverdata.mean_over_speed_perc, c = colormap[model.labels_], s = 40)
plt.title('KMeans Cluster')
plt.show()
```

```python
sse = []
list_k = list(range(1, 10))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(dd)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```
